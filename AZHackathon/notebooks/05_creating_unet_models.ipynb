```python
import torch
import torch.nn as nn
from torchvision.models import resnet152
```

```python
class UNet3p(nn.Module):

    def __init__(self, in_channels):
        super().__init__()
        self.backbone = resnet152(pretrained=False)
        
        self.backbone.conv1 = nn.Conv2d(
            in_channels,
            self.backbone.conv1.out_channels,
            kernel_size = self.backbone.conv1.kernel_size,
            stride = self.backbone.conv1.stride,
            padding = self.backbone.conv1.padding,
            bias = self.backbone.conv1.bias
        )
        print(self.backbone)
        print(list([*self.backbone.layer1]))
    
        #print(backbone)
        #backbone.layer1
        #backbone.layer2
        #backbone.layer3
        #backbone.layer4
        #print(self.backbone)
        
    def forward(self, x):
        print(x.shape)
        x = self.backbone.conv1(x)
        x = self.backbone.bn1(x)
        x = self.backbone.relu(x)
        x = self.backbone.maxpool(x)

        for i in range(len(self.backbone.layer1)):
            print(x.shape)
            x = self.backbone.layer1[i](x)
        for i in range(len(self.backbone.layer2)):
            print(x.shape)
            x = self.backbone.layer2[i](x)
        for i in range(len(self.backbone.layer3)):
            print(x.shape)
            x = self.backbone.layer3[i](x)
        for i in range(len(self.backbone.layer4)):
            print(x.shape)
            x = self.backbone.layer4[i](x)

        return x
model = UNet3p(1)

inputs = torch.rand(1,1,256,256)
outputs = model(inputs)
outputs.shape
```
