```python
import sys
sys.path.insert(0, "../../src")

from torch.utils.data import DataLoader

from data.augmentations import affine_augmentations, test_augmentations
from data.dataset import ExampleDataset
```

```python


train_dataset = ExampleDataset("../../data/03_training_data/normalized_bias/train", transform=affine_augmentations())
#train_dataset = ExampleDataset("../../data/02_intermediate/", transform=affine_augmentations())
train_dataloader = DataLoader(train_dataset, batch_size=1, num_workers=0)

```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_style("darkgrid")

_, axs = plt.subplots(2,7, figsize=(20,6))
for i, x in enumerate(inputs[0]):

    axs[0,i].imshow(x)
    
    hist, freq = np.histogram(x, bins=100)
    axs[1,i].plot(freq[1:], hist)
    
_, axs = plt.subplots(2,3, figsize=(20,12))
for i, x in enumerate(targets[0]):
    axs[0,i].imshow(x)
    hist, freq = np.histogram(x, bins=100)
    axs[1,i].plot(freq[1:], hist)
   
_, axs = plt.subplots(2,3, figsize=(20,12))
for i, x in enumerate(masks[0]):
    axs[0,i].imshow(x)
    hist, freq = np.histogram(x, bins=100)
    axs[1,i].plot(freq[1:], hist)
```
